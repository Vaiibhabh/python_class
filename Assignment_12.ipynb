# 2. From a PdfFileReader object, how do you get a Page object for page 5?

Ans. Calling getPage(4) will return a Page object for page 5 since page 0 is the first page

```python
import PyPDF2 as pdf
pdfFileObj = open("paper.pdf",'rb')
pdfReader1 = pdf.PdfReader(pdfFileObj)
pageObj = pdfReader1.pages[1]
pageObj.extract_text()
```

# 6. What is the difference between a Run object and a Paragraph object?

Paragraph Object: A document has many paragraphs. There are numerous runs in a paragraph, which starts on a new line. A collection of Paragraph objects representing the paragraphs in the document are contained in the Document object. (While typing in a Word document, pressing ENTER or RETURN causes a new paragraph to start.)

Run object: Runs are consecutive groups of characters in a paragraph that share the same style.

# 11. How do you add a paragraph with the text 'Hello, there!' to a Document object stored in a variable named doc?

```python

import docx
doc = docx.Document()

doc.add_paragraph('Hello there!')
doc.save('hellothere.docx')
```
